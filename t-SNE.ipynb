```python
import json
import numpy as np
```

```python
data = []
with open('en-train.json','r') as fp:
    for f in fp:
        data.append(json.loads(f))
```

```python
data[0:2]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = [d['text'] for d in data]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(corpus)

labels = [d['label'] for d in data]

labels = np.asarray(labels)
```

```python
sum(labels)/len(labels)
```

```python
from sklearn.manifold import TSNE
```

```python
X_embedded = TSNE(n_components=2, learning_rate='auto',
...                   init='random', perplexity=3).fit_transform(X)
```

```python
X_embedded.shape
```

```python
import matplotlib.pyplot as plt
```

```python
list_filter = lambda l,v : np.where(l == v)[0]
```

```python
plt.scatter(X_embedded[list_filter(labels, 0),0], X_embedded[list_filter(labels, 0),1])
plt.scatter(X_embedded[list_filter(labels, 1),0], X_embedded[list_filter(labels, 1),1])
plt.legend(['not protest','protest'])
plt.title('t-SNE visualization from TF-IDF weighting on en-train data.')
plt.savefig('en-tfidf-tsne.png')
plt.show()
```
