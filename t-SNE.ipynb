```python
import json
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.manifold import TSNE
```

```python
data = []
with open('en-train.json','r') as fp:
    for f in fp:
        data.append(json.loads(f))
```

```python
data[0:2]
```

```python
corpus = [d['text'] for d in data]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(corpus)

labels = [d['label'] for d in data]

labels = np.asarray(labels)
```

```python
sum(labels)/len(labels)
```

```python
X_embedded = TSNE(n_components=2, learning_rate='auto',
...                   init='random', perplexity=30).fit_transform(X)
```

```python
X_embedded.shape
```

```python
list_filter = lambda l,v : np.where(l == v)[0]
```

```python
plt.scatter(X_embedded[list_filter(labels, 0),0], X_embedded[list_filter(labels, 0),1])
plt.scatter(X_embedded[list_filter(labels, 1),0], X_embedded[list_filter(labels, 1),1])
plt.legend(['not protest','protest'])
plt.title('t-SNE visualization from TF-IDF weighting on en-train data.')
plt.savefig('en-tfidf-tsne.png')
plt.show()
```

```python
df = pd.DataFrame(X_embedded, columns = ['dim-0','dim-1'])

plt.figure(figsize=(16,10))
legend_map = {0: 'not protest',
              1: 'protest',
              }
fig = sns.scatterplot(
    x='dim-0',y='dim-1',
    # x="pca-one", y="pca-two",
    hue= np.vectorize(legend_map.__getitem__)(labels),
    palette=sns.color_palette("dark", 2),
    data=df,
    legend='full',
    alpha=0.75
)
plt.title('t-SNE visualization from TF-IDF weighting on en-train data.')
plt.savefig('sns_en-tfidf-tsne.png')
plt.show()
```
